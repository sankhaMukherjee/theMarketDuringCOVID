```python
import os, json
from tqdm import tqdm_notebook as tqdm
import financeMacroFactors as fM
from datetime import datetime as dt
from datetime import timedelta as tDel
from time import sleep
```

# Get company list

Now we shall download the stock data. This is similarly easy. We shall first get the ticker symbols for which we have fundamental data:

```python
files = [f for f in os.listdir('fundData') if f.endswith('.json')]
tickers = [f.replace('.json', '').lower() for f in files]
len(tickers)
```

# Download stock data

Next the stock data is downloaded from Yahoo Finance. This data is similarly stashed into JSON files to keep things simple. Ideally you would want to create a database and save this data.

Monthly averaged stock data is downloaded and saved into JSON files for all the tickers.

```python
for fileName in tqdm(files):
    sleep(0.5)
    
    ticker = fileName.replace('.json', '').lower()
    priceData  = fM.companies.getStockDataYahoo(
                    ticker, 
                    dt.now()-tDel(365*7), dt.now()+tDel(60), 
                    convert=False)

    json.dump( priceData, open(f'stockData/{ticker.upper()}.json', 'w') )
```
