```python
import financeMacroFactors as fM
import json
from tqdm.notebook import tqdm
```

# Get the Ticker Symbols

For downloading the data for the SNP 500 companies, we must frist download the ticker symbols of the companies in the S&P 500 list. Since we have already cached this data, this can be directly loaded form the cache.

```python
tickers = [c['Symbol'] for c in json.load(open('data/companies.json'))]
tickers[:10]
```

# Download the Fundamental data

Once the ticker symbols are known, the fundamental data may be downloaded and cached locally. It is ideal to save this data in a database. However, for simplicity, this data will be downloaded into JSON files for use later.

```python
for t in tqdm(tickers):
    print(f'------[{t:5s}]-------------')
    data = fM.companies.getTickerFundamentalDataMW(t)
    json.dump( data, open(f'fundData/{t}.json', 'w') )
```
